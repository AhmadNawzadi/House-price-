```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import set_config
import seaborn as sns
set_config(display='diagram')
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import LinearSVR
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, RepeatedKFold, train_test_split, cross_val_score
sns.set()
```

```python
# importer les fichier de train et test
df_train = pd.read_csv('/Users/ahmadnawzadi/machine learning/regression/train.csv')
df_test = pd.read_csv('test.csv')
df_train.head()
```

```python
# la correlation entre les features 
plt.figure(figsize = (30,30))
cor = df_train.corr()
sns.heatmap(cor, annot=True, cmap="PiYG")
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('House Price Distribution Plot')
sns.histplot(df_train.SalePrice, color ='red')

plt.subplot(1,2,2)
plt.title('House Price Spread')
ax = sns.boxplot(y = df_train.SalePrice, color = 'pink' )

plt.show()
```

```python
# verifier les valeurs nulls 
null_values = df_train.isna().sum().sort_values(ascending=False)
null_values.head(20)
```

```python
# supprimer des colunms contenant trop de Null valeur 
df_train = df_train.drop(columns=['Alley','PoolQC','MiscFeature','Fence','FireplaceQu'])
df_test = df_test.drop(columns=['Alley','PoolQC','MiscFeature','Fence','FireplaceQu'])
```

```python
# Pipeline + transformation + preprocessor

num_transformer = Pipeline(steps=[('imputer', SimpleImputer(
    strategy='mean')),('scaler', StandardScaler())])

cat_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', 
                fill_value='missing')),('onehot', OneHotEncoder(handle_unknown='ignore'))])

df_cat = df_train.select_dtypes(include = "object").columns
df_num = df_train.select_dtypes(exclude = "object").drop(['SalePrice'],axis=1).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, df_num),
        ('cat', cat_transformer, df_cat),
    ])
```

```python
# determiner le X et y
X = df_train.drop('SalePrice',axis=1)
y = df_train.SalePrice
```

```python
#split data set on test and train data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

### Le Model SVR

```python
# le model SVR
svr = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', LinearSVR())])
```

```python
#parameters for pipeline model
paramsLinearSVR = {
 'regressor__C': [float(i)*0.1+0.1 for i in range(10)],
 'regressor__max_iter': [ 1000*float(i+9) for i in range(7)],
 'regressor__tol': [0.001],
}
```

```python
#cross validation
cv1 = RepeatedKFold(n_splits=10, n_repeats=4, random_state=1)

# RandomizedSearchCV pour trouver des meilleur parameters
Grid = RandomizedSearchCV(svr, paramsLinearSVR, n_iter = 20, scoring='r2', cv=cv1)
Grid.fit(X_train, y_train) 
best_model = Grid.best_estimator_
```

```python
y_pred = best_model.predict(X_test)

# Show metrics
mse = mean_squared_error(y_test, y_pred)
mse = np.sqrt(mse)
mse = mean_squared_error(np.log(y_test), np.log(y_pred), squared=False)
print("MSE:", mse)

# print("RMSE:", rmse)
r2 = r2_score(y_test, y_pred)
print("R2:", r2)
```

### Le Model Ridge

```python
# Create the Ridge model
from sklearn.linear_model import Ridge
rig = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', Ridge(alpha=100))])
```

```python
rig.fit(X_train, y_train)
y_pred = rig.predict(X_test)
mse = mean_squared_error(np.log(y_test), np.log(y_pred), squared=False)
print('mse:',mse)

r2 = r2_score(y_test, y_pred)
print('r2:',r2)
```
